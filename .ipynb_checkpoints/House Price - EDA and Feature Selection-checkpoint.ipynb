**3.3. Dealing with Missing Values**

```python
# Calculate the missing data
print('House Pricing Features | % Missing values')
print('-'*41)

missing_values = data_house.isna().sum()[data_house.isna().sum() != 0]
number_of_row = data_house.shape[0]
missing_percentage = round(missing_values * 100 / number_of_row, 2)
print(missing_percentage.sort_values(ascending=False))
```

**Observation**
- Here, there are 4 columns that mostly has the missing values, with missing values more than 
  70%, which is 'PoolQC', 'MiscFeature', 'Alley', 'Fence'.
- It means that 4 columns will not give any impact to the model because most of the cell is missing. So, that we could actually   get rid of this 4 columns from our datasets.
- Columns with more than 50% of missing values wil be dropped

```python
# drop the columns with more 50% missing values
data_house.drop(['PoolQC', 'MiscFeature', 'Alley', 'Fence','Id'], inplace=True, axis=1)
data_house.shape
```

```python
data_house.drop(['FireplaceQu'], inplace=True, axis=1)
data_house.shape
```

**3.4. Identify Correlated Variable**

```python
# Correlation of each numerical columns
corr_data = data_house.corr()
corr_data
```

```python
# Heatmap correlation
plt.figure(figsize=(12,12))
sns.heatmap(corr_data, vmax = 1, vmin = -1, cmap = 'Blues', square = True)
```

```python
# Correlation of 'SalesPrice' compare to other numerical column
corr_data['SalePrice'].plot(kind = 'bar', figsize = (12,8), title = 'The Correlation between SalePrice and other Attributes')
plt.show()
```

**Observation**
- From the graph above, there are lot of columns that has less correlation to the SalePrice
- Next step, we just consider all the columns that has high impact to the SalePrice and drop the rest of the columns
- Here, we use Pareto Principle approach, which said that 80% of result comes from 20% effort. So, we choose all the columns that will impact to or correlate 0.80 with respect to SalePrice.

```python
high_corr_columns = list(corr_data.corr()['SalePrice'][(corr_data.corr()['SalePrice'] >= 0.80)
                                                            | 
                                                            (corr_data.corr()['SalePrice'] <= -0.80)].index)
high_corr_columns
```

```python
# New data that only contains all columns that has high impact to 'SalePrice'
new_data_house = data_house[high_corr_columns]
new_data_house.shape
```

```python
new_data_house.head()
```

```python
corr_new_data = new_data_house.corr()
plt.figure(figsize=(12,12))
sns.heatmap(corr_new_data, vmax=1, vmin=-1, cmap='Blues', square=True)
```

```python
corr_new_data['SalePrice'].plot(kind = 'bar', figsize = (12,8), title = 'The Correlation between SalePrice and other Attributes')
plt.show()
```

```python
# Heatmap for missing values
plt.figure(figsize=(8,6))
sns.heatmap(new_data_house.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Missing value
new_data_house.isna().sum()
```

**Observation**
- Here, there are one column that still has a missing values or null values.
- From above, there are 8 rows in column 'MasVnrArea' which is still missing.
- We need to decide what action that needs to be taken to tackle missing values, wheather fill the missing with 0, drop all the   rows that is missing, or fill missing value with Measure of Central Tendency
- We need to detect the outliers first

```python
# Boxplot of MasVnrArea
plt.figure(figsize=(10,6))
sns.boxplot(new_data_house['MasVnrArea'])
```

```python
# Distribution Plot
plt.figure(figsize=(10,6))
sns.distplot(new_data_house['MasVnrArea'], kde=True, color='blue', bins=30)
```

**Obeservation**
- According to the boxplot and distribution plot, this graphic is left-skewened
- The options is to fill missing values with either median or mean value
- Due to the skewened distribution happened, it is better for us to choose the median value. It is reasonable because mean value   will tend to the left according to the distribution plot. 

```python
# Fill missing values with medien
new_data_house['MasVnrArea'].fillna(new_data_house['MasVnrArea'].median(), inplace=True)
new_data_house.isna().sum()
```
